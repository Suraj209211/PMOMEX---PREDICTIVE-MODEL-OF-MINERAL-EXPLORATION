```
import geopandas as gpd
import os
import pandas as pd
import numpy as np
```

```
def read_shapefiles(directories):
    shapefiles = []
    for directory in directories:
        shapefiles.extend([os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.shp')])
    geodataframes = [gpd.read_file(shp) for shp in shapefiles]
    return geodataframes

shapefiles_directories = ['/content/drive/MyDrive/PREDICTIVE MODULE OF MINERAL EXPLORATION/PMOMEX---PREDICTIVE-MODEL-OF-MINERAL-EXPLORATION/DATA/RAW DATA/RAJASTHAN/GEOCHEMISTRY DATA/soil_regolith_gcs_ngdr_45K02',
                          '/content/drive/MyDrive/PREDICTIVE MODULE OF MINERAL EXPLORATION/PMOMEX---PREDICTIVE-MODEL-OF-MINERAL-EXPLORATION/DATA/RAW DATA/RAJASTHAN/GEOCHEMISTRY DATA/soil_regolith_gcs_ngdr_45K03',
                          '/content/drive/MyDrive/PREDICTIVE MODULE OF MINERAL EXPLORATION/PMOMEX---PREDICTIVE-MODEL-OF-MINERAL-EXPLORATION/DATA/RAW DATA/RAJASTHAN/GEOCHEMISTRY DATA/soil_regolith_gcs_ngdr_45K04',
                          '/content/drive/MyDrive/PREDICTIVE MODULE OF MINERAL EXPLORATION/PMOMEX---PREDICTIVE-MODEL-OF-MINERAL-EXPLORATION/DATA/RAW DATA/RAJASTHAN/GEOCHEMISTRY DATA/soil_regolith_gcs_ngdr_45K05',
                          ]
geodataframes = read_shapefiles(shapefiles_directories)
```

```
def clean_geodataframe(gdf):
    gdf = gdf.dropna()
    gdf = gdf.drop_duplicates()
    gdf = gdf[gdf.is_valid]
    return gdf

cleaned_geodataframes = [clean_geodataframe(gdf) for gdf in geodataframes]
```

```
merged_geodataframe = gpd.GeoDataFrame(pd.concat(cleaned_geodataframes, ignore_index=True))
```

```
import os

output_directory = '/content/drive/MyDrive/PREDICTIVE MODULE OF MINERAL EXPLORATION/PMOMEX---PREDICTIVE-MODEL-OF-MINERAL-EXPLORATION/CODE/DATA-PREPROCESSING'

def save_shapefile(gdf, filename):
    # Ensure the output directory exists
    if not os.path.exists(output_directory):
        print(f"Output directory does not exist: {output_directory}")
        return

    # Construct the full output path within the function
    output_path = os.path.join(output_directory, filename)

    try:
        gdf.to_file(output_path)
        print(f"Saved shapefile: {output_path}")
    except Exception as e:
        print(f"Error saving shapefile {output_path}: {e}")

for idx, gdf in enumerate(cleaned_geodataframes):
    save_shapefile(gdf, f'chemprocessed_shapefile_{idx}.shp')
```
